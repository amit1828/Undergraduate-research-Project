```python
import os
import pandas as pd

# List to store data from all CSVs
merged_data = []

# Get all CSV files in the current directory
csv_files = [file for file in os.listdir('.') if file.endswith('.csv')]

# Loop through each CSV file and extract the required columns
for file in csv_files:
    try:
        # Read the CSV file
        df = pd.read_csv(file)
        
        # Check if the required columns exist
        if "YS(MPa)" in df.columns and "HV(GPa)" in df.columns:
            # Select the relevant columns
            selected_data = df[["YS(MPa)", "HV(GPa)"]]
            
            # Append to the merged list
            merged_data.append(selected_data)
        else:
            print(f"Columns not found in {file}. Skipping.")
    except Exception as e:
        print(f"Error processing {file}: {e}")

# Concatenate all the data into a single DataFrame
if merged_data:
    merged_df = pd.concat(merged_data, ignore_index=True)

    # Save the merged data to a new CSV file
    merged_df.to_csv('all.csv', index=False)
    print("Merged data saved as 'merged_yield_strength_hardness.csv'.")
else:
    print("No valid data found to merge.")
```
