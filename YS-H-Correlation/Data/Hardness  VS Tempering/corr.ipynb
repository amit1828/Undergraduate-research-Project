```python
# Import the pandas library
import pandas as pd

# Read the CSV file (ensure "1.csv" is in your current working directory or provide the correct path)
df = pd.read_csv('1.csv')

# Display the DataFrame to verify the contents
print("DataFrame:")
print(df)

# Calculate correlation between the third column 'Hardness(HV1)' and the first column 'Tempering Time(Hrs)'
corr_time = df['Hardness(HV1)'].corr(df['Tempering Time(Hrs)'])
print("\nCorrelation between Hardness(HV1) and Tempering Time(Hrs):", corr_time)

# Calculate correlation between the third column 'Hardness(HV1)' and the second column 'Tempering Temp(C)'
corr_temp = df['Hardness(HV1)'].corr(df['Tempering Temp(C)'])
print("Correlation between Hardness(HV1) and Tempering Temp(C):", corr_temp)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the CSV file
df = pd.read_csv('1.csv')

# Prepare the predictor features and target variable
X = df[['Tempering Time(Hrs)', 'Tempering Temp(C)']]
y = df['Hardness(HV1)']

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict Hardness(HV1)
df['Predicted_Hardness'] = model.predict(X)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data points
ax.scatter(df['Tempering Time(Hrs)'], df['Tempering Temp(C)'], df['Hardness(HV1)'], color='blue', label='Actual Hardness')

# Scatter plot of predicted data points
ax.scatter(df['Tempering Time(Hrs)'], df['Tempering Temp(C)'], df['Predicted_Hardness'], color='red', label='Predicted Hardness')

# Set labels and title
ax.set_xlabel('Tempering Time (Hrs)')
ax.set_ylabel('Tempering Temp (C)')
ax.set_zlabel('Hardness (HV1)')
ax.set_title('3D Plot: Actual vs Predicted Hardness')
ax.legend()

plt.show()
```
